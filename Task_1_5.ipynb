Перевести ваш спарсенный json в формат csv таблицу через pandas 

```python
from traitlets.config.loader import filefind
import pandas as pd
import re
import json
with open ('/content/data.json', mode='r') as file:
    json_data = json.load(file)
json_data=json_data['data']
keys = json_data[0].keys()

df = pd.DataFrame(json_data)
df.to_csv('hh_python.csv', sep='\t')
df.insert(3,'salary rub',[0.0]*len(df),True)
df.insert(2,'swork experience int',[0.0]*len(df),True)
for index, row in df.iterrows():
    if row['work experience'] == 'Требуемый опыт работы: 1–3 года':
        df.at[index,'swork experience int']=2.0
    elif row['work experience'] == 'Требуемый опыт работы: 3–6 лет':
        df.at[index,'swork experience int']=4.5
    elif row['work experience'] == 'Требуемый опыт работы: более 6 лет':
        df.at[index,'swork experience int']=10.0
    df.at[index,'salary rub']=None
    if ('бел. руб.' in row['salary'] and ('до' in row['salary'] or 'от' in row['salary'])):
        st= ''.join(re.findall('\d',row['salary']))   
        df.at[index,'salary rub']=float(st)*23.36
    elif ('руб.' in row['salary'] and '–' in row['salary']):
        st=row['salary']
        st=st.replace(' ','')
        st=st.replace('\u202f','')
        st1,st2= st.split('–') 
        st2=st2.replace('руб.','')
  #      print (st1,':',st2)
        df.at[index,'salary rub']=(float(st1)+float(st2))/2
    elif ('USD' in row['salary'] and '–' in row['salary']):
        st=row['salary']
        st=st.replace(' ','')
        st=st.replace('\u202f','')
        st1,st2= st.split('–') 
        st2=st2.replace('USD','')
  #      print (st1,':',st2)
        df.at[index,'salary rub']=(float(st1)+float(st2))/2*58.4485
    elif ('KZT' in row['salary'] and '–' in row['salary']):
        st=row['salary']
        st=st.replace(' ','')
        st=st.replace('\u202f','')
        st1,st2= st.split('–') 
        st2=st2.replace('KZT','')
  #      print (st1,':',st2)
        df.at[index,'salary rub']=(float(st1)+float(st2))/2*0.1198
    elif ('KZT' in row['salary'] and ('до' in row['salary'] or 'от' in row['salary'])):
        st= ''.join(re.findall('\d',row['salary']))   
        df.at[index,'salary rub']=float(st)*0.1198     
    elif ('USD' in row['salary'] and ('до' in row['salary'] or 'от' in row['salary'])):
        st= ''.join(re.findall('\d',row['salary']))   
        df.at[index,'salary rub']=float(st)*58.4485
    elif ('EUR' in row['salary'] and ('до' in row['salary'] or 'от' in row['salary'])):
        st= ''.join(re.findall('\d',row['salary']))   
        df.at[index,'salary rub']=float(st)*55.7232
    elif ('руб' in row['salary'] and ('до' in row['salary'] or 'от' in row['salary'])):
        st= ''.join(re.findall('\d',row['salary']))   
        df.at[index,'salary rub']=float(st)
    if (',' in row['region']):
        df.at[index,'region']=row['region'].split(',')[0]   
df
```

Сгруппируйте вакансии по направлениями (DS, DE Software Engenering, etc.)
Какая средняя и медианная зарплата по группам вакансий?

```python

agg ={'salary rub':['median','mean'],'title':['count']}
grouped=df.groupby('title').agg(agg).reset_index()
grouped=grouped.dropna()
grouped.sort_values([('title','count')],ascending=False).head(20)
```

Какая средняя и медианная зарплата по каждому региону?

```python
agg ={'salary rub':['median','mean'],'region':['count']}
grouped=df.groupby('region').agg(agg).reset_index()
grouped=grouped.dropna()
grouped.sort_values([('region','count')],ascending=False).head(20)
```

Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?

```python
agg ={'salary rub':['median','mean'],'title':['count']}
grouped=df.groupby('title').agg(agg).reset_index()
grouped=grouped.dropna()
grouped.sort_values([('salary rub','mean')],ascending=False).head(1)
```

Какое процентное соотношение каждого региона по вакансиям от всех вакансий?


```python
import math
agg ={'region':['count']}
grouped=df.groupby('region').agg(agg).reset_index()
grouped.insert(2,'procent',' ',True)
for index, row in grouped.iterrows():
    grouped.at[index,'procent']= str(round(row[1]*100/len(df),2))+'%'
grouped.sort_values([('region','count')],ascending=False).head(10)
```
